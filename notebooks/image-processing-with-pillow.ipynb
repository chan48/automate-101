```python
import shutil
import os

from PIL import Image
```

```python
if "cat" in os.listdir("./"):
    shutil.rmtree("cat")
os.makedirs("cat")
```

```python
image = Image.open("./cat.jpg")
```

```python
# 확장자 변경하기

image.save("./cat/cat.png")
image.save("./cat/cat.tiff")
```

```python
from PIL import ImageFilter
```

```python
image.filter(ImageFilter.BLUR)
```

```python
from PIL import ImageFont
from PIL import ImageDraw
```

```python
draw = ImageDraw.Draw(image)
```

```python
draw
```

```python
font = ImageFont.truetype("./fonts/NanumBarunGothic.ttf", 16)
```

```python
font
```

```python
draw.text(
    (0, 0),
    "cat.png",
    (255, 255, 255),
    font=font,
)
```
